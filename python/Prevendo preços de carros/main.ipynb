01 - Importando e Criando Database

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn import preprocessing

df = pd.read_csv("carros_olx.csv")
df.head(2)
```

02 - Informações

```python
basic_info = pd.DataFrame({
    "Features": df.columns,
    "Missing Values": df.isnull().sum().values,
    "Missing Values %": ((df.isnull().sum().values)/len(df)),
    "Unique Values": df.nunique().values,
    "Data Types": df.dtypes
})
print(f'Linhas, Colunas = {df.shape}')
basic_info.reset_index(drop=True)
```

03 - Limpando Dados

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df.drop(['Ad ID','Car Name','Description',"Images URL's",'Car Profile'], axis=1, inplace=True)

print(f'Linhas, Colunas = {df.shape}')
df.head()
```

04 - Criando Modelo Preditivo

```python
y = df['Price']
X = df[list(df.columns.values)]

le = preprocessing.LabelEncoder()
for column_name in X.columns:
    if X[column_name].dtype == 'object':
        X[column_name] = le.fit_transform(X[column_name])
    else:
        pass

car_model = DecisionTreeRegressor(random_state=1)
```

05 - Teste de Treino

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)

car_model.fit(train_X, train_y)

prediction = car_model.predict(val_X)
print(f'Previsões = {prediction}')
```

06 - Calculando Margem de Erro

```python
mae = mean_absolute_error(val_y, prediction)
print(f'Margem de Erro: {mae}')

print(f'Valores reais = {list(val_y.head())}')
print(f'Valores previstos = {prediction[:5]}')
```
